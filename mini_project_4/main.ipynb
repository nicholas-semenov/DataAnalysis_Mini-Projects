```python
# Formulation of hypotheses

#Null hypothesis (H0): There is no significant difference in delivery time between experimental and control groups.
#Alternative hypothesis (H1): There are significant differences in delivery time between experimental and control groups.
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as ss
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(16,6)}, style='whitegrid')
```

```python
#Data loading and preparation
```

```python
df = pd.read_csv('experiment_lesson_4.csv')
df.head(4)
```

```python
print(df.shape)
df.info()
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
print(df['experiment_group'].value_counts())
```

```python
grouped = df.groupby('experiment_group')

group_old = grouped.get_group('control')
group_new = grouped.get_group('test')

# Пример вывода результатов
print("A group with an old algorithm:")
print(group_old)

print("\nA group with a new 'on the way' algorithm:")
print(group_new)
```

```python
print(group_old.describe())
print('\n')
print(group_new.describe())
```

```python
#Data visualisation
```

```python
# Visualisation of the distribution of delivery times for each group
plt.figure(figsize=(10, 6))
sns.histplot(group_old['delivery_time'], kde=True, color='blue', label='Control')
sns.histplot(group_new['delivery_time'], kde=True, color='orange', label='Experiment')
plt.legend()
plt.title('Distribution of Delivery Time by Group')
plt.xlabel('Delivery Time (minutes)')
plt.ylabel('Frequency')
plt.show()

# Boxplot for a visual comparison
plt.figure(figsize=(10, 6))
sns.boxplot(x='experiment_group', y='delivery_time', data=df)
plt.title('Boxplot of Delivery Time by Group')
plt.xlabel('Experiment Group')
plt.ylabel('Delivery Time (minutes)')
plt.show()
```

```python
# Descriptive statistics on districts and groups
district_grouped_data = df.groupby(['district', 'experiment_group'])['delivery_time'].describe()
print(district_grouped_data)

# Visualisation by neighbourhood
plt.figure(figsize=(14, 8))
sns.boxplot(x='district', y='delivery_time', hue='experiment_group', data=df)
plt.title('Boxplot of Delivery Time by District and Group')
plt.xlabel('District')
plt.ylabel('Delivery Time (minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Statistical test - t-test
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(group_old['delivery_time'], group_new['delivery_time'], equal_var=False)
print(f"T-Statistic: {t_stat}, P-Value: {p_value}")

# Checking the significance of the results
alpha = 0.05
if p_value < alpha:
    print("Differences are statistically significant")
else:
    print("Differences are not statistically significant")
```
